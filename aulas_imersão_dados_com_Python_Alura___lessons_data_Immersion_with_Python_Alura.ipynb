```python
columns_translation = {
    'work_year': 'ano',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_funcionario',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=columns_translation)
print("Colunas traduzidas:")
print(df.columns)
```

```python
df['nivel_experiencia'].value_counts()
```

```python
df['tipo_emprego'].value_counts()
```

```python
df['taxa_remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
experience_level_translation = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['nivel_experiencia'] = df['nivel_experiencia'].replace(experience_level_translation)
print("Categorias de 'nivel_experiencia' traduzidas:")
print(df['nivel_experiencia'].value_counts())
```

```python
employment_type_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['tipo_emprego'] = df['tipo_emprego'].replace(employment_type_translation)
print("Categorias de 'tipo_emprego' traduzidas:")
print(df['tipo_emprego'].value_counts())
```

```python
remote_ratio_translation = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['taxa_remoto'] = df['taxa_remoto'].replace(remote_ratio_translation)
print("Categorias de 'taxa_remoto' traduzidas:")
print(df['taxa_remoto'].value_counts())
```

```python
company_size_translation = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(company_size_translation)
print("Categorias de 'tamanho_empresa' traduzidas:")
print(df['tamanho_empresa'].value_counts())
```

```python
import numpy as np

#Criação de um dataframe de teste
df_salarios = pd.DataFrame({
    'nomes': ['Ana', 'Daniel', 'Samuel', 'Rute', 'Vinícius'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#Calcula a média, substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''
Calcula a mediana e substitui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas =pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()

df_temperaturas
```

```python
df_temperaturas =pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'Nomes': ['Ana', 'Daniel', 'Samuel', 'Rute', 'Vinícius'],
    'Cidade': ['São Paulo', np.nan, 'Nova Friburgo', np.nan, 'Bom Jardim']
})

df_cidades['cidade_preenchida'] = df_cidades['Cidade'].fillna('Não Informado')

df_cidades
```

```python
import plotly.express as px
```

```python
df_avg_salary = df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().reindex(ordem_senioridade).reset_index()
fig = px.bar(
    df_avg_salary,
    x='nivel_experiencia',
    y='salario_em_usd',
    title='Média Salarial Anual por Nível de Experiência',
    labels={'nivel_experiencia': 'Nível de Experiência', 'salario_em_usd': 'Média Salarial Anual (USD)'},
    color='nivel_experiencia',
    color_discrete_sequence=px.colors.qualitative.Set2
)
fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':ordem_senioridade})
fig.show()
```

```python
remoto_contagem = df_limpo['taxa_remoto'].value_counts().reset_index()
remoto_contagem.columns = ['Tipo_trabalho', 'Quantidade']
fig = px.pie(
    remoto_contagem,
    names = 'Tipo_trabalho',
    values = 'Quantidade',
    title='Proporção dos tipos de trabalho',
    )
fig.show()
```

# Desafio Aula 3 | Gráfico Interativo da Média Salarial dos Cientistas de Dados por País (Versão 1)

# Lesson 3 Challenge | Interactive Chart of Data Scientists Average Salaries by Country (Version 1)

```python
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
df_avg_salary_country = df_data_scientist.groupby('localizacao_empresa')['salario_em_usd'].mean().reset_index()
df_avg_salary_country_sorted = df_avg_salary_country.sort_values(by='salario_em_usd', ascending=False).head(20)

fig = px.bar(
    df_avg_salary_country_sorted,
    x='localizacao_empresa',
    y='salario_em_usd',
    title='Top 20 Países com Maior Média Salarial Anual para Cientistas de Dados',
    labels={'localizacao_empresa': 'País da Empresa', 'salario_em_usd': 'Média Salarial Anual (USD)'},
    color='localizacao_empresa'
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

# Desafio Aula 3 | Gráfico Interativo da Média Salarial dos Cientistas de Dados por País (Versão 2)

# Lesson 3 Challenge | Interactive Chart of Data Scientists Average Salaries by Country (Version 2)

```python
pip install pycountry
```

```python
import pycountry

def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2 = code).alpha_3
  except:
    return None

df_limpo['residencia_funcionario_iso3'] = df_limpo['residencia_funcionario'].apply(iso2_to_iso3)

df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_funcionario_iso3')['salario_em_usd'].mean().reset_index()

fig = px.choropleth(media_ds_pais,
                    locations = 'residencia_funcionario_iso3',
                    color = 'salario_em_usd',
                    color_continuous_scale = 'rdylgn',
                    title = 'Salário médio de Data Scientists por país',
                    labels = {'salario_em_usd': 'Salário médio (USD)', 'residencia_funcionario_iso3': 'País'})

fig.show()
```
